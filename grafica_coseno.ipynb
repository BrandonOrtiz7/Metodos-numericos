**GRAFICA DEL COSENO**
**Y=COS(X)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x=np.linspace(-2,4 * np.pi, 400) #especificamos el rango de valores
```

```python
y=np.cos(x) #calcula el valor de la funcion coseno
```

```python
plt.plot(x, y, label="cos(x)") #grafica la funcion con label asignamos el nombre cos a esa funcion graficada
plt.title("GRAFICA DEL COSENO") #titulo de la grafica
plt.xlabel("x") #nombre del eje x
plt.ylabel("cos(x)") #nombre del eje y
plt.legend() #muestra la leyenda
plt.grid()
plt.show() #muestra la grafica
```
